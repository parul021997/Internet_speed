```python
from tkinter import *
from speedtest import Speedtest

#GUI-------------------


ms = Tk()
ms.geometry("1000x700")
ms.title("Speed Test")
ms.resizable(width=False, height=False)
st = Speedtest()


frame2 = Frame(ms, bd=5, bg="aquamarine4", highlightthickness=3, highlightbackground="black", height=110, width=1000)
frame2.pack(side=TOP)
label = Label(ms, text="Internet Speed Test", bg="aquamarine4", font="Calibri 40 bold")
label.place(x=300, y=1)


frame=Frame(ms,bg='aquamarine3',highlightbackground='aquamarine3',highlightthickness=3,height=600, width=1000)
frame.pack()

frame1=Frame(frame,bg='aquamarine3',highlightbackground='aquamarine3',highlightthickness=3,height=600, width=1000)
frame1.place(x=500,y=1)


v1 = IntVar()
c1 = Checkbutton(ms, text="Download Speed", variable=v1, onvalue=1, offvalue=0)
c1.place(x=250, y=150)
v2 = IntVar()
c2 = Checkbutton(ms, text="Upload Speed", variable=v2, onvalue=1, offvalue=0)
c2.place(x=250, y=200)
v3 = IntVar()
c3 = Checkbutton(ms, text="Download Speed in  Megabits ", variable=v3, onvalue=1, offvalue=0)
c3.place(x=250, y=250)
v4 = IntVar()
c4 = Checkbutton(ms, text="Upload Speed in Megabits", variable=v4, onvalue=1, offvalue=0)
c4.place(x=250, y=300)
v5 = IntVar()
c5 = Checkbutton(ms, text="Download Speed in Megabytes", variable=v5, onvalue=1, offvalue=0)
c5.place(x=250, y=350)
v6 = IntVar()
c6 = Checkbutton(ms, text="Upload Speed in Megabytes", variable=v6, onvalue=1, offvalue=0)
c6.place(x=250, y=400)
f = Frame(frame1, bd=5, bg="white", highlightthickness =3, highlightbackground="black", height=250, width=450)
f.place(x=2, y=30)



def exit():
    ms.destroy()


def reset():
    c1.deselect()
    c2.deselect()
    c3.deselect()
    c4.deselect()
    c5.deselect()
    c6.deselect()
    test1.destroy()
    test2.destroy()
    


def calculateSpeedTest():
    print("Checking...")
    #For Checking Upload and Download
    dn = st.download()
    up = st.upload()
    print("Your Download Speed====>",dn,"Kb/s")
    print("Your Upload Speed=======>",up,"Kb/s")

    #Now Convert Data into Megabits --
    print("Your Download Speed in Megabits====>",round(dn/10**6,2),"Mb/s")
    print("Your Upload Speed in Megabits=======>",round(up/10**6,2),"Mb/s")

    #Speed ====> 7Mb/s == 1MB/s
    #Now Convert  Data into MegaBytes---
    print("Your Download Speed in MegaBytes====>",(dn/10**6)/7,"MB/s")
    print("Your Upload Speed in MegaBytes=======>",(up/10**6)/7,"MB/s")

    #Also Calculate Ping---
    # Get ping (miliseconds)
    ping = st.results.ping
    print("Ping in Milliseonds===>",ping,"/millisecond")
    print("Ping in seonds===>",ping/10**3,"/second")

    result = ""
    if(v1.get() == 1):
        result += 'Download Speed {} Kb/s \n'.format(dn)
    if(v2.get() == 1):
        result += 'Upload Speed {} Kb/s \n'.format(up)
    if(v3.get() == 1):
        result += 'Download Speed in  Megabits {} Mb/s \n'.format(round(dn/10**6,2))
    if(v4.get() == 1):
        result += 'Upload Speed in  Megabits {} Mb/s \n'.format(round(up/10**6,2))
    if(v5.get() == 1):
        result += 'Download Speed in Megabytes {} MB/s \n'.format((dn/10**6)/7)
    if(v6.get() == 1):
        result += 'Upload Speed in Megabytest {} MB/s \n'.format((up/10**6)/7)


    global test1
    test1 = Label(f,
                  text= result,
                  bg="white", font="Calibri 10 bold", justify=LEFT,height=13)
    test1.place(x=2, y=2)



button1 = Button(ms, text="Exit", bg='aquamarine2', bd=15, command=exit)
button1.place(x=400,y=500)
button2 = Button(frame1, text="Speed", bg="aquamarine2", bd=15, command=calculateSpeedTest)
button2.place(x=100, y=385)
button3 = Button(frame, text="Reset", bg="aquamarine2", bd=15, command=reset)
button3.place(x=170, y=390)



ms.mainloop()
```
